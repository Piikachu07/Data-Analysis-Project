# Key Insights
1. There is -1 in **Salary Estimate**, **Rating**, **Founded** and in **Competitors** so i have to addresses this during the data cleaning.
2. The dataset has diverse job titles, company sizes, industries, and locations.
Some columns have dominant values, such as **Data Scientist** in Job Title and Unknown / Non-Applicable in Revenue.
3. Outliers in numerical data, particularly in **Rating** and **Founded**, need addressing

```python
df.nunique()
```

```python
df_cleaned = df.copy()

salary = df_cleaned['Salary Estimate'].str.extract(r'(\$\d+(K|M)?-\$\d+(K|M)?)' )[0]

# Split the salary range into min and max salary
df_cleaned[['Min_salary', 'Max_salary']] = salary.str.split('-', expand=True)

# Removed "$" and "K" from both columns and convert to numeric
df_cleaned['Min_salary'] = df_cleaned['Min_salary'].str.replace(r'[\$K]', '', regex=True).astype(float) * 1000
df_cleaned['Max_salary'] = df_cleaned['Max_salary'].str.replace(r'[\$K]', '', regex=True).astype(float) * 1000


df_cleaned['Average_salary'] = (df_cleaned['Min_salary'] + df_cleaned['Max_salary'])/2

Location_split = df['Location'].str.split(',', expand = True)

df_cleaned['city'] = Location_split[0]
df_cleaned['State'] = Location_split[1]
df_cleaned['Company Name'] = df_cleaned['Company Name'].str.split(r'\n' ).str[0]

df_cleaned[df_cleaned['Rating']< 0 ].fillna(df_cleaned['Rating'].mean(), inplace=True)
df_cleaned = df_cleaned[df_cleaned['Founded']> 0 ]
df_cleaned = df_cleaned[df_cleaned['Rating']> 0 ]

df_cleaned.drop(columns=['Unnamed: 0', 'Salary Estimate'], inplace=True)

df_cleaned

```

```python
# Chart 1: Distribution of Ratings
plt.figure(figsize=(8, 5))
sns.histplot(df_cleaned['Rating'], bins=20, kde=True, color='purple')
plt.title("Distribution of Company Ratings", fontsize=14)
plt.xlabel("Rating")
plt.ylabel("Frequency")
plt.show()
```

Businesses can use this insight to benchmark their ratings against competitors and focus on improving employee satisfaction to enhance their reputation

```python
# Chart-2: Count of Jobs by Industry

# Get the top N industries
top_n = 15  # Change this number to include more or fewer industries
top_industries = df_cleaned["Industry"].value_counts().nlargest(top_n).index

# Filter the DataFrame to include only these industries
filtered_df = df_cleaned[df_cleaned["Industry"].isin(top_industries)]

# Create the count plot
plt.figure(figsize=(12, 6))
sns.countplot(y=filtered_df["Industry"], order=top_industries, palette="coolwarm")
plt.title("Number of Job Listings by Top Industries", fontsize=14)
plt.xlabel("Count", fontsize=12)
plt.ylabel("Industry", fontsize=12)
plt.show()
```

By analyzing this graph job seekers can target industries with higher demand, while businesses in low-demand industries can strategize ways to attract talent.

```python
# Chart-3: Number of Job Listings by Company

top_companies = df_cleaned["Company Name"].value_counts().nlargest(15)  # Top 15 companies by grouping them

plt.figure(figsize=(12, 6))
sns.barplot(x=top_companies.values, y=top_companies.index, palette="viridis")
plt.title("Top 15 Companies with Most Job Listings", fontsize=14)
plt.xlabel("Number of Listings", fontsize=12)
plt.ylabel("Company Name", fontsize=12)
plt.show()
```

job seekers can prioritize applications to companies with higher hiring activity, and companies can assess their hiring competitiveness.

```python
# chart 13- Industries with the highest average rating
# Compute average rating by industry and select the top 15
avg_rating_industry = df.groupby("Industry")["Rating"].mean().sort_values(ascending=False).head(15)

# Plot the bar chart
plt.figure(figsize=(12, 6))
sns.barplot(x=avg_rating_industry.values, y=avg_rating_industry.index, palette="coolwarm")

# Titles and labels
plt.title("Top 15 Industries by Average Company Rating", fontsize=14)
plt.xlabel("Average Rating", fontsize=12)
plt.ylabel("Industry", fontsize=12)

# Show the chart
plt.show()

```

```python
# chart-14 How Competitors affects the actively hiring
def count_competitors(x):
    if x == "-1" or pd.isna(x):
        return 0
    else:
        # Split by comma and count
        return len(x.split(','))

# Create a new column for the number of competitors
df["Num_Competitors"] = df["Competitors"].apply(count_competitors)

plt.figure(figsize=(10, 6))
sns.histplot(df["Num_Competitors"], bins=np.arange(0, df["Num_Competitors"].max()+2)-0.5, color="teal", kde=False)
plt.title("Distribution of Number of Competitors", fontsize=14)
plt.xlabel("Number of Competitors", fontsize=12)
plt.ylabel("Frequency", fontsize=12)
plt.show()
```

```python
# Chart 15-> Onweship of the companies in the top 5 sectors 
# Count of Companies by Ownership Type Across Top 5 Sectors
# Identify top 5 sectors by number of listings
top5_sectors = df_cleaned["Sector"].value_counts().nlargest(5).index
df_top5 = df_cleaned[df_cleaned["Sector"].isin(top5_sectors)]

plt.figure(figsize=(12, 6))
sns.countplot(x="Sector", hue="Type of ownership", data=df_top5, palette="Set1")
plt.title("Count of Companies by Ownership Type in Top 5 Sectors", fontsize=14)
plt.xlabel("Sector", fontsize=12)
plt.ylabel("Count", fontsize=12)
plt.xticks(rotation=45, ha="right")
plt.legend(title="Type of Ownership")
plt.show()
```
